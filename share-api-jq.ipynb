# SHARE API data work
I exported the simple query 'ucsd' as a JSON object. This notebook will use `jq` to query and filter it. 

```python
DATA="share_api/share_api01.json"
```

Simple count of results:

```python
!cat $DATA | jq '.count'
```

Let's look at it all, prettified:

```python
!cat $DATA | jq '.'
```

```python
!cat $DATA | jq '.results'
```

Get all the titles:

```python
!cat $DATA | jq '.results[].title'
```

Start getting at the source info:

```python
!cat $DATA | jq '.results[].shareProperties'
```

Let's get it down to just sources:

```python
!cat $DATA | jq '.results[].shareProperties.source'
```

Now let's only get unique values:

```python
!cat $DATA | jq '[.results[].shareProperties.source] | unique'
```
